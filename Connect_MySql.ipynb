```python
import mysql.connector

# Establish connection
try:
    mydb = mysql.connector.connect(
        host="localhost",
        user="root",
        password="Rudra33077",
        database="test"
    )
    print("Database connection successful!")
except mysql.connector.Error as err:
    print(f"Error connecting to database: {err}")
    mydb = None # Set mydb to None if connection fails
```

```python
# Consistent table name and safe insert for 'mystudents' table
if mydb is not None:
    cursor = mydb.cursor()
    try:
        # Create the 'students' table if it doesn't exist
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS mystudents (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(50),
            age INT,
            course VARCHAR(50)
        )
        """)

        # Insert data into 'mystudents' table
        sql = "INSERT INTO mystudents (name, age, course) VALUES (%s, %s, %s)"
        val = ("Rudra Pratap Singh", 23, "MCA")
        cursor.execute(sql, val)
        mydb.commit()   # Save the changes
        print(cursor.rowcount, "record inserted successfully.")
    except mysql.connector.Error as err:
        print("MySQL error:", err)
else:
    print("No database connection. Ensure credentials are correct.")
```

```python
# Retrieve data
cursor.execute("SELECT * FROM mystudents")
result = cursor.fetchall()

print("\nStudent Records:")
for row in result:
    print(row)

# Close connection
mydb.close()
```
