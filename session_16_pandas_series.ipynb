## üîπ What is Pandas Made Up Of?

Pandas is built **on top of NumPy** ‚Äî it uses NumPy arrays internally for fast computation,
but adds **labels** and **axes (rows & columns)** to make data easier to handle.

So:

```
NumPy ‚Üí Fast number crunching
Pandas ‚Üí Organized, labeled data
```

## üîπ Main Pandas Objects (Data Structures)

Pandas provides **three main objects** to study and use:

| Object        | Description                        | Similar To                              |
| ------------- | ---------------------------------- | --------------------------------------- |
| **Series**    | 1-D labeled array                  | Column in Excel / 1 column in DataFrame |
| **DataFrame** | 2-D labeled table (rows + columns) | Excel sheet / SQL table                 |
| **Index**     | Labels for rows or columns         | Row/column headers                      |

---

### 1Ô∏è‚É£ **Series**

* One-dimensional (like a column).
* Has both **values** and **index labels**.

Example:

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s)
```

 Output:

```
a    10
b    20
c    30
d    40
dtype: int64
```

* You can access by index:

```python
print(s['b'])   # 20
```

---

### 2Ô∏è‚É£ **DataFrame**

* Two-dimensional table (rows √ó columns).
* Each column is a **Series**.

Example:

```python
data = {
    'Name': ['Rudra', 'Amit', 'Sneha'],
    'Age': [22, 24, 21],
    'Marks': [85, 90, 95]
}

df = pd.DataFrame(data)
print(df)
```

‚úÖ Output:

```
    Name  Age  Marks
0  Rudra   22     85
1   Amit   24     90
2  Sneha   21     95
```

* Access columns:

```python
print(df['Name'])    # Series
```

* Access rows:

```python
print(df.loc[1])     # by label/index
```

---

### 3Ô∏è **Index**

* Defines the **labels of rows and columns** in a Series or DataFrame.
* Can be custom or auto-generated.

Example:

```python
print(df.index)      # RangeIndex(start=0, stop=3, step=1)
print(df.columns)    # Index(['Name', 'Age', 'Marks'], dtype='object')
```
