````python
# import titanic
df = pd.read_csv('/content/titanic.csv')
df['Name']
````

````python
# applying regex
# contains
# question: search the name which contain john in frist name, either upper case or lower
# search john -> both case
df[df['firstname'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')] # for vowel
# df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')] # for consonent
````

# ⭐ **PANDAS STRING OPERATIONS – Study Notes**

Pandas provides powerful **vectorized string functions** using the **`.str` accessor**.

---

# 🔵 **1. What Are Vectorized Operations?**

**Vectorized operations** allow you to apply an operation on **entire arrays/columns at once**, instead of looping through items one-by-one.

### ✔ Example (vectorized)

```python
df['Name'].str.upper()
```

Pandas applies `.upper()` to all rows **in optimized C code**, not Python loops.

---

# 🔵 **2. Problem in Vanilla Python (No Vectorization)**

In Python lists, strings must be processed with **loops**:

```python
[x.upper() for x in names]
```

Problems:

* Slow (Python loops are interpreted, not optimized)
* More code
* Hard to apply advanced functions
* Cannot handle missing values (NaN)
* Inconsistent behavior

---

# 🔵 **3. How Pandas Solves This Issue?**

Pandas uses:

* **Vectorization**
* **NumPy backend**
* **`.str` accessor**
* Built-in handling for **NaN**
* Regex support

`.str` internally broadcasts functions efficiently on entire columns.

Example:

```python
df['Name'].str.lower()
```

Fast, clean, works with NaN.

---

# 🔵 **4. String Accessor – `.str`**

Used to access **vectorized string functions**.

```python
df['City'].str
```

You must use `.str` before calling string methods:

```python
df['City'].str.upper()
```

---

# 🔵 **5. COMMON STRING FUNCTIONS (Most Used)**

## ✔ lower() / upper() / title() / capitalize()

```python
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.title()
df['Name'].str.capitalize()
```

---

## ✔ len()

Length of each string.

```python
df['Name'].str.len()
```

---

## ✔ strip(), lstrip(), rstrip()

Removes spaces.

```python
df['Name'].str.strip()
```

---

## ✔ split()

Split into list.

```python
df['FullName'].str.split(" ")
```

Extract first/last name:

```python
df['FullName'].str.split().str[0]      # First name
df['FullName'].str.split().str[-1]     # Last name
```

---

## ✔ get()

Returns element at a particular position.

```python
df['FullName'].str.split().str.get(1)
```

---

## ✔ replace()

```python
df['City'].str.replace("Delhi", "New Delhi")
```

Supports regex by default.

---

# 🔵 **6. FILTERING Using String Patterns**

## ✔ startswith()

```python
df[df['Name'].str.startswith("A")]
```

## ✔ endswith()

```python
df[df['Name'].str.endswith("n")]
```

## ✔ contains()

```python
df[df['Name'].str.contains("john")]
```

---

# 🔵 **7. Case-Insensitive Search (Find 'john' or 'John' or 'JOHN')**

```python
df[df['Name'].str.contains("john", case=False)]
```

---

# 🔵 **8. Regex Filtering (Powerful)**

Search names with only letters:

```python
df[df['Name'].str.contains("^[A-Za-z]+$", regex=True)]
```

Find all rows where name has digits:

```python
df['Name'].str.contains(r"\d")
```

---

# 🔵 **9. Slicing String Columns**

Use `.str` + slice:

```python
df['Code'].str[:3]     # first 3 characters
df['Code'].str[2:]     # from index 2 to end
df['Code'].str[-2:]    # last 2 characters
```

---

# 🔵 **10. Combining Multiple String Functions**

Example:

```python
df['Name'].str.strip().str.lower().str.replace(" ", "_")
```

---

# 🔵 **11. Important `.str` Functions Summary**

| Category   | Functions                                |
| ---------- | ---------------------------------------- |
| Case       | lower, upper, title, capitalize          |
| Whitespace | strip, rstrip, lstrip                    |
| Splitting  | split, rsplit, get                       |
| Matching   | contains, startswith, endswith           |
| Replacing  | replace                                  |
| Length     | len                                      |
| Slicing    | `str[ start:end ]`                       |
| Regex      | contains(regex), replace(regex), extract |

---

# ⭐ **Quick Revision (One-Liners)**

* `.str` enables **vectorized string operations**
* Vanilla Python loops are slow → Pandas uses NumPy backend
* `.str` handles **missing values** automatically
* Use `case=False` for case-insensitive search
* `.str.contains()` and `.str.replace()` fully support **regex**
* `.str[ ]` allows slicing
* Used heavily in **cleaning**, **feature engineering**, and **text preprocessing**
