# why separate objects to handle data and time when python already has datetime functionality?

````python
# why separate objects to handle data and time when python already has datetime functionality?
````

- syntax wise datetime is very convenient
- But the performance takes a hit while working with huge data. List vs Numpy Array
- The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
- The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.

````python
import numpy as np
date = np.array('2015-07-04', dtype=np.datetime64)
date
````

````python
date + np.arange(12)
````

- Because of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large

- Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64

- From a group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or DataFrame

#  Date and Time in Pandas  Study Notes

Pandas has special types to work with dates & times:

* **`Timestamp`**  single date-time value
* **`DatetimeIndex`**  index made of many timestamps
* **`to_datetime()`**  convert data to datetime
* **`date_range()`**  generate a sequence of dates

---

## 1 Creating a `Timestamp`

### Basic creation

```python
import pandas as pd

t = pd.Timestamp('2025-01-15')
```

### Different variations

```python
pd.Timestamp(2025, 1, 15)                 # year, month, day
pd.Timestamp('2025/01/15')                # different string format
pd.Timestamp('15-01-2025', dayfirst=True) # day first
```

### With time (24-hour format)

```python
pd.Timestamp('2025-01-15 14:30:45')       # 2:30:45 PM
```

### With AM/PM

```python
pd.Timestamp('2025-01-15 2:30 PM')
pd.Timestamp('2025-01-15 02:30:00 PM')
```

### Using `datetime.datetime` object

```python
import datetime as dt

dt_obj = dt.datetime(2025, 1, 15, 14, 30, 0)
t = pd.Timestamp(dt_obj)
```

---

## 2 Fetching Attributes from `Timestamp`

Given:

```python
x = pd.Timestamp('2025-01-15 14:30:45')
```

You can access:

```python
x.year      # 2025
x.month     # 1
x.day       # 15
x.hour      # 14
x.minute    # 30
x.second    # 45
```

Also:

```python
x.day_name()   # 'Wednesday' etc.
x.date()       # date only (Python date)
x.time()       # time only (Python time)
```

---

## 3 Why Separate Objects When Python Already Has `datetime`?

Pandas adds features **on top of** Pythons `datetime`:

* Works efficiently on **entire columns** (vectorized)
* Handles **missing values** (`NaT`, like `NaN` for time)
* Powerful **time-series index** (`DatetimeIndex`)
* **Frequency handling** (`freq='D'`, `'M'`, `'H'`, etc.)
* Easy **resampling**, **shifting**, **rolling windows**
* Can use with `Series` and `DataFrame` directly

So, Pandas date/time is designed for **data analysis**, not just representing a single date-time.

---

## 4 `DatetimeIndex` Object

A `DatetimeIndex` is an index made of timestamps.

### From strings

```python
dates = pd.to_datetime(['2025-01-01', '2025-01-05', '2025-01-10'])
idx = pd.DatetimeIndex(dates)
```

### Using Python `datetime` objects

```python
import datetime as dt

dates = [dt.datetime(2025, 1, 1), dt.datetime(2025, 1, 5)]
idx = pd.DatetimeIndex(dates)
```

### Using `pd.Timestamp`

```python
dates = [pd.Timestamp(2025, 1, 1), pd.Timestamp(2025, 1, 5)]
idx = pd.DatetimeIndex(dates)
```

### Using as Series index

```python
s = pd.Series([10, 20], index=idx)
```

Now you can do:

```python
s['2025-01-01']
s['2025-01']      # all data in Jan 2025
```

## 5 `date_range()`  Generating Date Sequences

### Daily dates in a given range

```python
rng = pd.date_range(start='2025-01-01', end='2025-01-10', freq='D')
```

### Generate specific number of periods

```python
rng = pd.date_range(start='2025-01-01', periods=5, freq='D')
```

### Alternate days in a range (every 2 days)

```python
rng = pd.date_range(start='2025-01-01', end='2025-01-10', freq='2D')
```

---

## 6 Frequency (`freq`) Attribute  Common Values

Some common `freq` strings:

* `'D'`   Day
* `'2D'`  Every 2 days
* `'B'`   Business day
* `'H'`   Hour
* `'T'` or `'min'`  Minute
* `'S'`   Second
* `'W'`   Week
* `'M'`   Month end
* `'MS'`  Month start
* `'Q'`   Quarter end
* `'Y'`   Year end

Example:

```python
rng = pd.date_range('2025-01-01', '2025-03-01', freq='W')
rng.freq        # shows frequency
```

---

## 7 `to_datetime()`  Converting to Datetime

Used to convert strings / numbers / mixed formats into **datetime**.

### Basic usage

```python
pd.to_datetime(df['date_column'])
```

### With different formats

```python
pd.to_datetime(df['date_column'], dayfirst=True)
```

### Handling errors  `errors` parameter

* `'raise'`   default, error if bad data
* `'coerce'`  invalid values become `NaT`
* `'ignore'`  return original data if conversion fails

```python
pd.to_datetime(df['date_column'], errors='coerce')
```

This is very useful when data is **messy**.

---

## 8 Using `DatetimeIndex` with `Series` / `DataFrame`

```python
df = pd.DataFrame({'value': [10, 20, 30]},
                  index=pd.date_range('2025-01-01', periods=3, freq='D'))
```

Now you can:

```python
df['2025-01-02']       # select by date
df['2025-01']          # all in January
df.loc['2025-01-01':'2025-01-03']   # date range
```
