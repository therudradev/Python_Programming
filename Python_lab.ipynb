# Program: Find prime numbers between 1 and 100

````python
print("Prime numbers between 1 and 100 are:")

for num in range(2, 101):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

# Exp 9
# Program: To-Do List Management

````python
# Create an empty list to store tasks
tasks = []

while True:
    print("\n===== TO-DO LIST MENU =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        task = input("Enter the task to add: ")
        tasks.append(task)
        print(f"Task '{task}' added successfully!")

    elif choice == '2':
        if len(tasks) == 0:
            print("No tasks in the To-Do List.")
        else:
            print("\nYour To-Do List:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")

    elif choice == '3':
        if len(tasks) == 0:
            print("No tasks to remove.")
        else:
            print("\nCurrent Tasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")
            remove_index = int(input("Enter the task number to remove: "))
            if 0 < remove_index <= len(tasks):
                removed = tasks.pop(remove_index - 1)
                print(f"Task '{removed}' removed successfully!")
            else:
                print("Invalid task number.")

    elif choice == '4':
        print("Exiting To-Do List Program. Goodbye!")
        break

    else:
        print("Invalid choice! Please enter a number between 1 and 4.")

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

#Exp 11:
# Program: Count unique words in a given text

````python
# Take input from the user
text = input("Enter your text: ")

# Remove punctuation and symbols
clean_text = ""
for ch in text:
    if ch.isalnum() or ch.isspace():
        clean_text += ch

# Convert text to lowercase and split into words
words = clean_text.lower().split()

# Count unique words
unique_words = set(words)

# Display results
print(f"\nTotal words: {len(words)}")
print(f"Unique words: {len(unique_words)}")
print("\nUnique words list:")
print(unique_words)

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

# Program: Count frequency of elements in a list using dictionary

````python
# Create a list of elements
elements = [10, 20, 10, 30, 20, 10, 40, 30, 50, 40, 20]

# Create an empty dictionary to store frequencies
frequency = {}

# Count occurrences of each element
for item in elements:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

print("List:", elements)
print("\nFrequency of elements:")
for key, value in frequency.items():
    print(f"{key}: {value}")

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

#Exp 15
# Program: Demonstrate Filter, Map, and Lambda functions

````python
# Create a list of numbers
numbers = [i for i in range(1,11)]

# Use filter() with lambda to get even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

# Use map() with lambda to square all numbers
squared_numbers = list(map(lambda x: x ** 2, numbers))

# Combine filter() and map() — square only even numbers
even_squares = list(map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, numbers)))

# Print results
print("Original List:", numbers)
print("Even Numbers (Filter):", even_numbers)
print("Squared Numbers (Map):", squared_numbers)
print("Squares of Even Numbers (Filter + Map):", even_squares)

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

#Exp 16:
# Program: Demonstrate OOPs Concepts in Python


````python
from abc import ABC, abstractmethod  # For abstraction

# Class demonstrating self and __init__
class Student:
    def __init__(self, name, age):  # __init__ acts as a constructor
        self.name = name
        self.age = age

    def display(self):  # self refers to the current instance
        print(f"Name: {self.name}, Age: {self.age}")


# Encapsulation: Using private variables and getters/setters
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


# Inheritance:
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")


# Polymorphism: Same method behaves differently
def animal_sound(animal):
    animal.speak()


# Abstraction: Hiding implementation details using abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


# ---------- Main Program ----------
# Self and __init__
s1 = Student("Rudra", 23)
s1.display()

# Encapsulation
acc = BankAccount(5000)
acc.deposit(1500)
print("Current Balance:", acc.get_balance())

# Inheritance & Polymorphism
dog = Dog()
animal_sound(dog)

# Abstraction
circle = Circle(5)
print("Area of Circle:", circle.area())

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

# Program: Replace missing values with column mean

````python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {
    'Maths': [85, 90, np.nan, 70, 95],
    'Science': [88, np.nan, 76, np.nan, 90],
    'English': [80, 78, 82, 85, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame with Missing Values:\n")
print(df)

# Replace missing values with column mean
df_filled = df.fillna(df.mean(numeric_only=True))

print("\nDataFrame after Replacing Missing Values with Mean:\n")
print(df_filled)

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````

# Exp: 19
# Program: Normalize a NumPy array using Z-score and Min-Max scaling

````python
import numpy as np

# Create a sample NumPy array
data = np.array([10, 20, 30, 40, 50, 60])
print("Original Array:\n", data)

# Z-score Normalization
# Formula: z = (x - mean) / std
mean = np.mean(data)
std_dev = np.std(data)
z_score_normalized = (data - mean) / std_dev
print("\nZ-score Normalized Array:\n", z_score_normalized)

# Min-Max Scaling
# Formula: x' = (x - min) / (max - min)
min_val = np.min(data)
max_val = np.max(data)
min_max_scaled = (data - min_val) / (max_val - min_val)
print("\nMin-Max Scaled Array:\n", min_max_scaled)

print("\n--------------------------------------")
print("Experiment done by: Rudra Pratap Singh")
print("Roll No.: 13247057")
````
