````python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
````

````python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
````

````python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
````


#  **PANDAS `apply()` FUNCTION**

The **`apply()` function** is used to apply a **custom function** to a **Series or DataFrame**.

It is one of the most powerful tools for data manipulation.

---

#  **1. `apply()` on a Series (Column-wise)**

Used to apply a function to **each value** in a column.

###  Example

```python
df['Marks'].apply(lambda x: x + 5)
```

###  What it does?

Adds **5** to every value in the `Marks` column.

---

#  **2. `apply()` on a DataFrame (Row-wise / Column-wise)**

###  Apply function to **each column**

```python
df.apply(sum)
```

###  Apply function to **each row**

```python
df.apply(lambda row: row['Maths'] + row['Science'], axis=1)
```

* `axis=0`  apply function **column-wise**
* `axis=1`  apply function **row-wise**

---

#  **3. Using Built-in Functions with `apply()`**

```python
df['Name'].apply(len)        # length of each name
df['Marks'].apply(abs)       # absolute value
df['Salary'].apply(float)    # convert type
```

---

#  **4. Apply Custom Function**

###  Create your own function

```python
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    else:
        return "C"

df['Grade'] = df['Marks'].apply(grade)
```

---

#  **5. Apply with Multiple Columns (Row-wise Computation)**

```python
df['Total'] = df.apply(lambda r: r['Maths'] + r['Science'], axis=1)
```

---

#  **6. `applymap()` vs `map()` vs `apply()` (Important Difference)**

| Function     | Works On            | Meaning                         |
| ------------ | ------------------- | ------------------------------- |
| `apply()`    | Series or DataFrame | Apply function on rows/columns  |
| `map()`      | Series only         | One-to-one mapping              |
| `applymap()` | DataFrame only      | Apply function to **each cell** |

Example:

```python
df.applymap(str.upper)
```

---

#  **7. When to Use `apply()`**

Use `apply()` when:

* You need **custom logic**
* Operation is not available as a vectorized Pandas/NumPy function
* You want to calculate using **multiple columns**

---

#  **8. When NOT to Use `apply()`**

Avoid `apply()` when:

* A built-in vectorized function exists (faster)
* Looping over large datasets (apply is slower)

Example:
Instead of:

```python
df['Marks'].apply(lambda x: x + 5)
```

Use:

```python
df['Marks'] + 5
```

---

#  **Summary (One-Liner Notes)**

* `apply()`  apply a function on a column or row
* Use `axis=1` for row-wise operations
* Great for custom logic
* Slower than vectorized operations
* Works on both DataFrame and Series

#  **SORTING IN PANDAS**

Pandas provides two main ways to sort data:

1. **`sort_values()`**  sort by *column values*
2. **`sort_index()`**  sort by *index (row labels)*

---

#  **1. Sorting by Column Values  `sort_values()`**

###  Syntax:

```python
df.sort_values(by='column_name')
```

###  Example:

```python
df.sort_values(by='Age')
```

Sorts rows in **ascending order** (default).

---

##  **Sort in Descending Order**

```python
df.sort_values(by='Age', ascending=False)
```

---

##  **Sort by Multiple Columns**

```python
df.sort_values(by=['Age', 'Marks'])
```

###  With different orders:

```python
df.sort_values(by=['Age', 'Marks'], ascending=[True, False])
```

---

##  **Sort a Series**

```python
s.sort_values()
```

---

#  **2. Sorting by Index  `sort_index()`**

Sorts rows based on index labels.

###  Example:

```python
df.sort_index()
```

###  Descending order:

```python
df.sort_index(ascending=False)
```

---

#  **3. Important Parameters (Useful for Exams)**

| Parameter     | Meaning                                   |
| ------------- | ----------------------------------------- |
| `by`          | Column(s) to sort on                      |
| `ascending`   | True/False for order                      |
| `inplace`     | If True  updates original DataFrame      |
| `na_position` | Where to place NaN: `'first'` or `'last'` |

###  Example with parameters:

```python
df.sort_values(by='Marks', ascending=False, na_position='first')
```

---

#  **4. Sorting with `inplace=True`**

Updates the DataFrame directly.

```python
df.sort_values(by='Age', inplace=True)
```

---

#  **5. Mixed Sorting Example (Very Common)**

```python
df.sort_values(
    by=['City', 'Marks'],
    ascending=[True, False]
)
```

* Sort by **City** AZ
* If same city, sort by **Marks** highlow

---

#  **Why Sorting is Important**

* Helps organize data for analysis
* Used in reporting, ranking, and filtering
* Essential for EDA and data cleaning

#  **PANDAS IMPORTANT FUNCTIONS**

---

#  **1. `value_counts()`**

Returns the **count of unique values** in a column (Series).

```python
df['City'].value_counts()
```

**Key points:**

* Sorted in descending order by default
* Good for frequency distribution
* Use `normalize=True` for percentages

```python
df['City'].value_counts(normalize=True)
```

---

#  **2. `rank()`**

Assigns a **rank** to each value in a column.

```python
df['Marks'].rank()
```

**Key points:**

* Default: average ranks for ties
* Use `method='min'` or `'max'` for different styles
* Use `ascending=False` for reverse ranking

---

#  **3. `sort_index()`**

Sort rows by **index labels**.

```python
df.sort_index()
df.sort_index(ascending=False)
```

---

#  **4. `set_index()`**

Replaces the current index with a **column**.

```python
df.set_index('ID')
```

**Key points:**

* Use `inplace=True` to update DataFrame
* Use `drop=False` to keep column

---

#  **5. `rename()` / rename index**

Rename columns or index labels.

```python
df.rename(columns={'old': 'new'})
df.rename(index={0: 'first'})
```

---

#  **6. `reset_index()`**

Converts index back to normal column.

```python
df.reset_index()
df.reset_index(drop=True)
```

---

#  **7. `unique()` & `nunique()`**

###  `unique()`  unique values

```python
df['City'].unique()
```

###  `nunique()`  count of unique values

```python
df['City'].nunique()
```

---

#  **8. `isnull()` / `notnull()` / `hasnans`**

Check for missing values.

```python
df.isnull()
df.notnull()
df['Marks'].hasnans
```

---

#  **9. `dropna()`**

Remove rows/columns with missing values.

```python
df.dropna()
df.dropna(axis=1)   # drop columns
```

**Key options:**

* `how='any'` (default)
* `how='all'`
* `subset=['col1', 'col2']`

---

#  **10. `fillna()`**

Fill missing values with a specific value.

```python
df.fillna(0)
df.fillna(df['Marks'].mean())
```

#  **11. `drop_duplicates()`**


Apply a function to a column or row.
