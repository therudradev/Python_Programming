````python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
````

````python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
````

````python
genres.apply(foo)
````

````python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
````

````python
genres.apply(rank_movie)
````

````python
# find normalized IMDB rating group wise
# normalization formulla
# x(noramalized)=(x - x.min)/ (x.max - x.min)

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
````

# ⭐ **PANDAS `groupby()`**

`groupby()` is used to **split data into groups**, apply operations, and then combine results.

This is the **Split → Apply → Combine** process.

---

# 🔵 **1. Creating Groups**

```python
groups = df.groupby('Category')
```

For multiple columns:

```python
df.groupby(['Category', 'City'])
```

---

# 🔵 **2. FIND TOTAL NUMBER OF GROUPS**

```python
len(groups)
```

---

# 🔵 **3. FIND ITEMS IN EACH GROUP → `size()`**

```python
groups.size()
```

---

# 🔵 **4. FIRST / LAST / NTH ITEM**

### ✔ First row of each group:

```python
groups.first()
```

### ✔ Last row of each group:

```python
groups.last()
```

### ✔ Nth item from each group:

```python
groups.nth(2)   # 3rd row because index starts from 0
```

---

# 🔵 **5. `get_group()` vs Filtering**

### ✔ Using groupby:

```python
groups.get_group('Furniture')
```

### ✔ Using filtering:

```python
df[df['Category'] == 'Furniture']
```

**Difference:**

* `get_group()` is faster and cleaner
* Filtering scans the entire DataFrame again

---

# 🔵 **6. `groups` Attribute**

Shows dictionary of **group label → row indexes**.

```python
groups.groups
```

Example:

```
{'Furniture': [0, 3, 5],
 'Technology': [1, 4, 7]}
```

---

# 🔵 **7. `describe()`**

Gives descriptive statistics **per group**.

```python
groups.describe()
```

---

# 🔵 **8. `sample()`**

Get **random samples** from each group.

```python
groups.sample(1)
```

---

# 🔵 **9. `nunique()`**

Number of **unique values** IN each group.

```python
groups['City'].nunique()
```

---

# 🔵 **10. Aggregation – `agg()` Method**

### ✔ Single aggregation

```python
groups['Sales'].agg('sum')
```

### ✔ Multiple aggregations (single column)

```python
groups['Sales'].agg(['sum', 'mean', 'max'])
```

### ✔ Aggregations using dictionary

```python
groups.agg({
    'Sales': ['sum', 'mean'],
    'Quantity': ['max', 'min']
})
```

### ✔ Aggregations using list

```python
groups['Sales'].agg([sum, min, max])
```

---

# 🔵 **11. LOOPING OVER GROUPS**

```python
for name, group in groups:
    print(name)
    print(group)
```

Outputs:

* `name` = group label
* `group` = DataFrame of that group

---

# 🔵 **12. SPLIT → APPLY → COMBINE**

This is how `groupby()` works:

### ✔ Split

DataFrame is divided into groups.

### ✔ Apply

Function applied to each group.

### ✔ Combine

Results merged into a new DataFrame.

---

# 🔵 **13. Using `apply()` with groupby**

### ✔ With built-in function:

```python
groups['Sales'].apply(sum)
```

### ✔ Custom function:

```python
groups.apply(lambda g: g['Sales'].max() - g['Sales'].min())
```

---

# 🔵 **14. GroupBy on Multiple Columns**

```python
df.groupby(['Category', 'Region'])['Sales'].sum()
```

Creates hierarchical grouping.

---

# 🔵 **15. IMPORTANT GROUPBY METHODS (Quick List)**

| Method               | Purpose                  |
| -------------------- | ------------------------ |
| `size()`             | Count rows per group     |
| `count()`            | Count non-missing values |
| `sum()`              | Sum                      |
| `mean()`             | Average                  |
| `min()` / `max()`    | Min/Max                  |
| `first()` / `last()` | First/Last row           |
| `nth()`              | nth row                  |
| `nunique()`          | Unique values count      |
| `sample()`           | Random sample            |
| `describe()`         | Summary stats            |
| `agg()`              | Multiple aggregations    |
| `apply()`            | Custom logic             |

---

# 🎯 **Short Summary (For Revision)**

* `groupby()` groups rows based on column values
* `len()` → number of groups
* `size()` → number of rows in each group
* `get_group()` → extract a group
* `agg()` → multiple operations
* Supports multiple columns grouping
* Built on "Split–Apply–Combine" strategy
* Use `apply()` for custom logic
